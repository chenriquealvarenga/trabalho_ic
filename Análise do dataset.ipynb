```python
import pandas as pd
from IPython.display import display

features = pd.read_csv("data/dengue_features_train.csv")
labels = pd.read_csv("data/dengue_labels_train.csv")

print("Features (5 primeiras linhas)")
with pd.option_context("max_columns", None):
    display(features.head(5))
    
print("Labels (5 primeiras linhas)")
with pd.option_context("max_columns", None):
    display(labels.head(5))
```

```python
print("Features (descricao)")
with pd.option_context("max_columns", None):
    display(features.describe())
    
print("Labels (descricao)")
with pd.option_context("max_columns", None):
    display(labels.describe())
```

```python
print("Features - numero de valores nulos por coluna")
display(features.isnull().sum())

print("Labels - numero de valores nulos por coluna")
display(labels.isnull().sum())
```

```python
data = features.merge(labels, on=("city","year","weekofyear"))
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
def plot_corr(df,size=10):
    corr = df.corr()
    corr = corr.applymap(abs)
    fig, ax = plt.subplots(figsize=(size, size))
    cax = ax.matshow(corr, vmin=0,vmax=1)
    fig.colorbar(cax)
    plt.xticks(range(len(corr.columns)), corr.columns, rotation="vertical");
    plt.yticks(range(len(corr.columns)), corr.columns);
    
print("Correlacao entre variaveis (0 = nenhuma correlacao, 1 = correlacao total)")
plot_corr(data)
```

```python
print("Correlacao entre variaveis na cidade sj")
plot_corr(data[data.city=="sj"])
```

```python
print("Correlacao entre variaveis na cidade iq")
plot_corr(data[data.city=="iq"])
```

```python
import numpy as np

def plot_corr_ranking(data, title=""):
    fig, ax = plt.subplots(figsize=(15,6))
    corr_ranking = data.corr()["total_cases"].sort_values()
    corr_ranking = corr_ranking[corr_ranking.index!="total_cases"]
    labels = corr_ranking.index
    y_pos = np.arange(len(labels))
    values = corr_ranking.values
    ax.barh(y_pos, values, align='center', color='green')
    ax.set_yticks(y_pos)
    ax.set_yticklabels(labels)
    ax.axes.set_xlim((-0.3,0.3))
    plt.title(title)
    plt.show()

plot_corr_ranking(data[data.city=="sj"], title="Cidade: SJ")
plot_corr_ranking(data[data.city=="iq"], title="Cidade: IQ")
```

```python
def plot_total_cases(df, group_column, title="", rotation_x="vertical", size=(15,3)):
    plt.figure(figsize=size)
    df = df.groupby(group_column).total_cases.sum()
    x = df.index
    y = df.values
    plt.title(title)
    plt.xticks(x, rotation=rotation_x)
    plt.plot(x,y)
    plt.show()

# plot_total_cases(data, "year", title="Total de casos por ano (duas cidades)")
plot_total_cases(data[data.city=="sj"], "year", title="Total de casos por ano (SJ)")
plot_total_cases(data[data.city=="iq"], "year", title="Total de casos por ano (IQ)")

# plot_total_cases(data, "weekofyear", title="Total de casos por semana do ano (duas cidades)", 
#                  rotation_x="horizontal")
plot_total_cases(data[data.city=="sj"], "weekofyear", title="Total de casos por semana do ano (SJ)", 
                 rotation_x="horizontal")
plot_total_cases(data[data.city=="iq"], "weekofyear", title="Total de casos por semana do ano (IQ)", 
                 rotation_x="horizontal")
```
